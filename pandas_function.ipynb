```python
#мержим датафреймы
abc = pd.DataFrame({
    'name': ['кабачок', 'помидор', 'огурец'],
    'abc': ['A', 'A', 'C']
})

xyz = pd.DataFrame({
    'name_xyz': ['кабачок', 'баклажан', 'огурец'],
    'xyz': ['X', 'Y', 'Z']
})
```

```python
#удаляем лишние элементы
abc_xyz = abc.merge(xyz, left_on='name', right_on='name_xyz', how='outer')
abc_xyz
```

```python
import numpy as np
```

```python
abc_xyz['name'] = np.where(abc_xyz['name'].isnull(), abc_xyz['name_xyz'], abc_xyz['name'])
abc_xyz
```

```python
#фильтрация данных
df.filter(like='ug', axis=1)
```

```python
shops = df['DR_Apt'].unique().tolist()
```

```python
#разбиваем датафрейм на несколько отдельных и сохраняем в папку в пазных файлах
for shop in shops:
    small_df = df[df['DR_Apt'] == shop]
    small_df.to_excel(f'new_folder/{shop}.xlsx', sheet_name=f'{shop}', index=False, encoding='1251')
```

```python
#загружаем датафрейм из фала
pd.read_excel('new_folder/2.xlsx')
```

```python
#загружаем несколько датафреймов из файлов и обьеденяем
import glob
```

```python
glob.glob('new_folder/*.xlsx')
```

```python
files = glob.glob('new_folder/*.xlsx')

dfs = []

for file_name in files:
    small_df = pd.read_excel(file_name)
    dfs.append(small_df)
```

```python
len(dfs)
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
#пробегаем по каждой строке и генерируем сообщения
for i, row in df.iterrows():
    s = f"insert into sales values ('{row['DR_Dat']}', '{row['DR_Kol']}')"
    print(s)
    # print(f'Номер строки: {i}, дата: {row["DR_Dat"]}')
    if i == 10:
        break
```
